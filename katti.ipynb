```python
#x = input()
#y = input()
x = "2-"
y = "13"
x = "-3"
y = "21"
GOAL = [['1', '2'], ['3', '-']]

def createlist(x, y):
    return [list(x), list(y)]

def find_blank(l):
    for i in range(2):
        for j in range(2):
            if l[i][j] == '-':
                return i, j

CW_ORDER  = [(0,1),(0,0),(1,0),(1,1)]
CCW_ORDER = list(reversed(CW_ORDER))

def step_cw(l):
    i, j = find_blank(l)
    idx = CW_ORDER.index((i, j))
    ni, nj = CW_ORDER[(idx + 1) % 4]
    l[i][j], l[ni][nj] = l[ni][nj], l[i][j]  # move '-' one step CW
    return l

def step_ccw(l):
    i, j = find_blank(l)
    idx = CCW_ORDER.index((i, j))
    ni, nj = CCW_ORDER[(idx + 1) % 4]
    l[i][j], l[ni][nj] = l[ni][nj], l[i][j]  # move '-' one step CCW
    return l

def move_cw(l):
    from copy import deepcopy
    board = deepcopy(l)
    turns = 0
    for _ in range(4):
        if board == GOAL:
            return turns
        board = step_cw(board)
        turns += 1
    return turns  # full cycle if needed

def move_ccw(l):
    from copy import deepcopy
    board = deepcopy(l)
    turns = 0
    for _ in range(4):
        if board == GOAL:
            return turns
        board = step_ccw(board)
        turns += 1
    return turns


def move(l):
    if l == GOAL: return 0
    return min(move_cw(l), move_ccw(l))

l = createlist(x, y)
print(createlist(x, y))
print(move(l))
```
